Change Boolean to String on Expert Labels and Newcomers Labels item.

```python
df_labels_classification['Expert Labels'] = df_labels_classification['Expert Labels'].map({True: 'True', False: 'False'})
```

```python
df_labels_classification['Newcomers Labels'] = df_labels_classification['Newcomers Labels'].map({True: 'True', False: 'False'})
```
