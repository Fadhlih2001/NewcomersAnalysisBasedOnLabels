Try Remove Digits on labels

```python
def remove(list):
	pattern = '[0-9]'
	list = [re.sub(pattern, '', i) for i in list]
	return list
```

```python
def clear(list):
	pattern = ' - '
	list = [re.sub(pattern, '', i) for i in list]
	return list
```

```python
remove_list = remove(mylist)
```

```python
print(remove(mylist))
```

```python
clear_list = clear(remove_list)
```

```python
df_list_labels_clear = pd.DataFrame(clear_list)
```

```python
df_list_labels_clear
```

```python
df_list_labels_clear.to_csv("List_of_Labels_Clear_No_Duplication.csv")
```
